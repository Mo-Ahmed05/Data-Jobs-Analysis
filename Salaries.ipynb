```python
import seaborn as sns
import matplotlib.pyplot as plt
import Data
```

```python
df = Data.DF.copy()
df.dropna(subset=['yearly_salary_avg'], inplace=True)
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist',
              'Senior Data Analyst', 'Senior Data Engineer', 'Senior Data Scientist']
df = df[df['job_title_short'].isin(job_titles)]

jobs_list = df.groupby('job_title_short')['yearly_salary_avg'].median().sort_values(ascending=False).index
```

```python
sns.set_theme()
plt.figure(figsize=(8,6))

sns.boxplot(data=df, y='job_title_short', x='yearly_salary_avg', order=jobs_list)
plt.xlabel('Yearly Salary (USD)')
plt.xlim(0, 600000)
plt.ylabel('')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))


plt.suptitle("Salary Distribution", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
